```python
import matplotlib.pyplot as plt

def plot_data(df_c):
    plt.figure(figsize=(20, 20))
    for i, col in enumerate(df_c.select_dtypes(include=['int', 'float'])):
        plt.subplot(5, 3, i + 1)
        df_c[col].plot(kind='hist', title=col)
        plt.xlabel(col)
```

```python
plot_data(df_c)
```

```python
def plot_data(df_a):
    plt.figure(figsize=(20, 20))
    for i, col in enumerate(df_a.select_dtypes(include=['int', 'float'])):
        plt.subplot(5, 3, i + 1)
        df_a[col].plot(kind='hist', title=col)
        plt.xlabel(col)
        
plot_data(df_a)
```

```python
numeric_df_a = df_a.select_dtypes(include=['number'])
corr_a = numeric_df_a.corr(method='pearson')
corr_a
```
